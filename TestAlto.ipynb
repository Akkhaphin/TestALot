```python
df = pd.read_csv('/content/drive/My Drive/intro_ml_workshop_dataset.csv')
df = df.set_index('ds')
df.index = pd.to_datetime(df.index)
```

```python
df.drop(['wetbulb_temperature', 'humidity'], axis=1, inplace=True)
df.head()
```

```python
train = df.loc[df.index < '2023-08-22']
test = df.loc[df.index >= '2023-08-22']
```

```python
from pycaret.regression import *
s = setup(data = train, test_data = test, fold= 3 , session_id= 123)
```
